### 导入相关库

```python
# from Rosmaster_Lib
import time
from Rosmaster_Lib import Rosmaster
# Camera
from lovon.camera_rosmaster import Rosmaster_Camera
```

### 运动指令下达测试

```python
# 创建Rosmaster对象 bot Create the Rosmaster object bot
bot = Rosmaster()
# 启动接收数据，只能启动一次，所有读取数据的功能都是基于此方法 
# Start to receive data, can only start once, all read data function is based on this method  
bot.create_receive_threading()
bot.set_car_motion(0, 0, 0)
```

### 测试IMU数据获取

```python
# 测试获取IMU数据测试
# bot.get_imu_attitude_data(ToAngle=True)
# 获取500次IMU数据，统计成功次数和失败次数
num_tests = 500
success_count = 0
failure_count = 0
for _ in range(num_tests):
    r, p, y = bot.get_imu_attitude_data(ToAngle=True)
    if r is not None and p is not None and y is not None:
        success_count += 1
    else:
        failure_count += 1
success_rate = success_count / num_tests * 100
print(f"Image Capture Success Rate: {success_rate}% ({success_count} successes, {failure_count} failures)")
```

### 运动指令下达测试

```python
# 测试50次运动测试
for _ in range(50):
    bot.set_car_motion(0.5, 0, 0)
    time.sleep(0.2)
    bot.set_car_motion(0, 0.5, 0)
    time.sleep(0.2)
    bot.set_car_motion(0, 0, 0.8)
    time.sleep(0.5)
    bot.set_car_motion(0, 0, 0)
```

### 图像获取测试

```python
# 获取500次图像，统计成功次数和失败次数
camera = Rosmaster_Camera(debug=False)
num_tests = 500
success_count = 0
failure_count = 0
for _ in range(num_tests):
    success, frame = camera.get_frame()
    if success:
        success_count += 1
    else:
        failure_count += 1
success_rate = (success_count / num_tests) * 100
print(f"Image Capture Success Rate: {success_rate}% ({success_count} successes, {failure_count} failures)")
```
